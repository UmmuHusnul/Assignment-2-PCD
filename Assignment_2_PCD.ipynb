### Imager Transform pada Citra Grayscale menggunakan metode Haar

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Baca citra Grayscale
image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Melakukan transformasi Haar pada citra Grayscale
def haar_transform(img):
    size = img.shape[0]
    while size >= 2:
        for i in range(0, size, 2):
            avg = (img[i] + img[i + 1]) / 2
            diff = (img[i] - img[i + 1]) / 2
            img[i] = avg
            img[i + 1] = diff
        size //= 2
    return img

haar_img = img.copy()
haar_img = haar_transform(haar_img)
```

```python
# Menampilkan citra asli dan citra yang telah diubah dengan transformasi Haar
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.title('Citra Asli')
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title('Hasil Transformasi Haar')
plt.imshow(haar_img, cmap='gray')
plt.show()
```
